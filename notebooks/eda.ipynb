## 1. Data Ingestion & Cleaning
We simulate the cleaning pipeline here to identify data anomalies like the Total Price POS truncation bug.

```python
# Example of the cleaning step
def check_pos_truncation_bug(df):
    # This bug divides Total Price by 10 or 100 on aggregate rows
    if 'Total Price' in df.columns and 'Total Cost' in df.columns and 'Total Profit' in df.columns:
        df['True Revenue'] = df['Total Cost'] + df['Total Profit']
        # Identify rows where reported Total Price is significantly lower than mathematically expected
        bug_rows = df[df['Total Price'] < df['True Revenue'] * 0.5]
        print(f'Found {len(bug_rows)} aggregate rows affected by the POS truncation bug.')
        return True
    return False
```
